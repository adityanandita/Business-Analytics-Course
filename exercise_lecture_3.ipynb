```python
def analysis_null_vale_table(table):
    null_value = table.isnull().sum()
    percentage_null_value = round(100*null_value/len(table),2)
    return pd.concat([null_value,percentage_null_value], axis=1).rename(columns={0:"Missing Count", 1:"Percentage (%)"})
```

```python
analysis_null_vale_table(data)
```

## **Data Cleaning**

```python
data.columns
```

```python
# Assume the No and Cust. ID are must.
check_column = data.columns.difference(['NO', 'Customer_ID'])
empty_column = data[check_column].isnull().all(axis=1)
print(data[empty_column]['Customer_ID'])
print(f'\nTotal data incompleted: {len(data[empty_column])}')
```

We know there are 25 accounts where the data is not completed. So We will delete this data from our dataframe.

```python
data_clean = data[~empty_column]
```

```python
# Then to tackle NaN value in job, we use this method:
data_clean['Job'] = data_clean['Job'].fillna('No Answer')

# To tackle there are no response, we use same method like Job
data_clean['Response'] = data_clean['Response'].fillna('No Answer')
```

So, the next step to increase our accuracy in analysis, we will try some advance analysis, there are:
- `Data cleaning` for Age
- `Data transformation` like log-transform for charge_Monday to reduce the influence of outliers and achieve a more normal distribution.
- Create `additional visualization` using boxplots to better understand the distribution and clearly identify outliers.

```python
# Data cleaning for Age
data_age = df_task_2[(df_task_2['Age'] >= 0) & (df_task_2['Age'] <= 120)]

# Data Transformation for charge_Monday
log_data = df_task_2
log_data['log_charge_Monday'] = np.log(log_data['charge_Monday'] + 1)
```

```python
# Histogram of log charge_Monday
sns.histplot(log_data['log_charge_Monday'], bins=30, kde=True)
plt.title('Histogram of Log charge_Monday')
plt.show()
```

```python
# Data Visualization using Boxplot

fig, axes = plt.subplots(1, 2, figsize=(14, 7))

sns.boxplot(x=data_age['Age'], ax=axes[0])
axes[0].set_title('Boxplot of Age Distribution')

sns.boxplot(x=log_data['log_charge_Monday'], ax=axes[1])
axes[1].set_title('Boxplot of Log-transformed charge_Monday Distribution')
plt.show()
```
