```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Read cp_data_eng_trial2.csv file as data
data = pd.read_csv('cp_data_eng_trial2.csv')

# Frequency Distribution for 'Response'
sns.countplot(x='Response', data=data)
plt.title('Frequency Distribution for Campaign Response')
plt.show()

# Frequency Distribution for 'Sex'
sns.countplot(x='Sex', data=data)
plt.title('Frequency Distribution for Sex')
plt.show()

# Frequency Distribution for 'Job'
sns.countplot(x='Job', data=data)
plt.title('Frequency Distribution for Job')
plt.show()
```
